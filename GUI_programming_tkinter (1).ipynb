#root.mainloop(): is a method on the main window which is executed when we run our application.
the method loops forever, waiting for events from the user, until the user exits the program either by closing the window, or by terminating the program with a keyboard interrupt in the console.

```python
#create a label as child of root window
from tkinter import Tk, Label
root=Tk()
msg=Label(root, text='welcome to Python programming lecture')
msg.pack()  #is used on each widget to position it inside its parent
root.mainloop()
```

```python
#to change font of the text
from tkinter import Tk, Label
root=Tk()
msg=Label(root, text='welcome to Python programming lecture')
msg.config(font=('callibri',24, 'italic bold underline'))
msg.pack()  #is used on each widget to position it inside its parent
root.mainloop()
```

```python
from tkinter import Tk, Label, Y, RIGHT
root=Tk()
label1=Label(root, text='welcome to Python programming lecture', background='red')
label2=Label(root, text='Have a good day', background='green')
label1.config(font=('callibri',24, 'italic bold underline'))
label2.config(font=('callibri',14, 'italic'))
label1.pack(fill=Y, padx=10, ipady=25, side=RIGHT)  #fill: widget expands to take up any extra space(X,Y, or BOTH)
label2.pack(fill=Y, padx=20, ipady=40, side=RIGHT) #padx, pady: outer padding, ipadx,ipady: inner padding
root.mainloop()
```

```python
#display two buttons and print a message when a button is clicked
from tkinter import Tk, Label, Button
root=Tk()
exitButton=Button(root, text='EXIT', command=root.destroy)
exitButton.pack()
def my_callback():
    print("Hi")
msg_button=Button(root, text="display message", command=my_callback)
msg_button.pack()
root.mainloop()
```

```python
#menu
import tkinter
from tkinter import Tk, Menu
root=Tk()
menu_bar=Menu(root)
fileMenu=Menu(menu_bar,tearoff=0) #tearoff=menu can popout
fileMenu.add_command(label="Stop", command=root.destroy)
fileMenu.add_command(label="kill", command=root.destroy)
menu_bar.add_cascade(label='File',menu=fileMenu) #Add fileMenu drop down as submenu of menu_bar
root.config(menu=menu_bar)
root.mainloop()
```

```python
#Input
import tkinter
from tkinter import Tk, Entry, Button, INSERT
root=Tk()
entry=Entry(root)
entry.pack()
def message():
    print(entry.get())
button=Button(root, text='print content', command=message)
button.pack()
root.mainloop()
```
